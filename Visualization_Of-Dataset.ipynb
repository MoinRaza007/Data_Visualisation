```python
rain = np.random.randint(5,15,30)
plt.plot(day,temperature,color="black",marker="o",linestyle = 'dotted',linewidth = 2)
plt.plot(day,rain,color="blue",marker="o",linestyle = '-',linewidth = 2)
plt.show()
```

```python
rain = np.random.randint(5,15,30)
plt.plot(day,temperature,color="black",marker="o",linestyle = 'dotted',linewidth = 2,label="Temp")
plt.plot(day,rain,color="blue",marker="o",linestyle = '-',linewidth = 2,label="Rain")
plt.legend()
plt.xlabel("Temp")
plt.ylabel("RAIN")
plt.title("Temp VS Rain")
plt.grid()
plt.show()
```

```python
## Scatterplot
import numpy as np
import pandas as pd
```

```python
rain = np.random.randint(5,15,30)
day = np.arange(1,31)
temperature = np.random.randint(10,25,30)

```

```python
df = (rain,day,temperature)
df
```

```python
data = list(zip(day,temperature,rain))
df = pd.DataFrame(data,columns=["Day","Temperature","Rain"])
df
```

```python
## syntax
plt.scatter(x="Temperature",y="Rain",data=df)
plt.xlabel("Temperature")
plt.ylabel("Rain")
plt.title('Temp VS Rain')
plt.grid()
plt.show()
```

```python
## syntax
plt.scatter(x="Temperature",y="Rain",data=df,alpha=.5)
plt.xlabel("Temperature")
plt.ylabel("Rain")
plt.title('Temp VS Rain')
plt.grid()
plt.show()
```

```python
### Distplot
```

```python
sns.distplot(df['petal_length'],hist=False)
```

```python
species = df.species.unique()
species
```

```python
for i in species:
  subset = df[df['species']==i]
  sns.distplot(subset['petal_length'],hist=False,kde=True,label=i)
plt.legend()
plt.show()
```
